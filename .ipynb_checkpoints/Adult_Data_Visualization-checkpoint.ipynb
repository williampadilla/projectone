```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# File to Load (Remember to Change These)
file_to_load = "adult.csv"

# Read Adult File and store into Pandas data frame
adult_data = pd.read_csv(file_to_load)
adult_data 

# check the dtypes

#adult_data.dtypes

# check for null values

adult_data.isnull().sum()

# Data cleaning

#for column in df:
    #print(column,df[column].unique())
    

adult_data['Occupation'].unique()

# remove data where occupation is ' ?'

df = adult_data.loc[adult_data['Occupation'] != ' ?'] 

df = df.loc[df['Workclass'] != ' Without-pay'] 

df['Marital_Status'].unique()
df['Marital_Status'].unique()

# remove columns that are insignificant for our analysis like fnlwgt, Capital-gain, Capital-loss
    
del df['fnlwgt']

del df['Capital-gain']

del df['Capital-loss']


# Replace Married-civ-spouse, Married-AF-spouse as Married
# Repalce Married-spouse-absent as Seperated


df['Marital_Status'] = df['Marital_Status'].replace([' Married-civ-spouse', ' Married-AF-spouse'], ' Married')

df['Marital_Status'] = df['Marital_Status'].replace([' Married-spouse-absent'], ' Separated')

df['Education'] = df['Education'].replace([' 11th',' 9th',' 7th-8th',' 5th-6th',' 10th',' Preschool',' 12th', ' 1st-4th'], ' Below-HS')


df['Education'] = df['Education'].replace([' Assoc-acdm', ' Assoc-voc'], ' Twoyear-Degree')

df['Workclass'] = df['Workclass'].replace([' State-gov',' Federal-gov',' Local-gov'], ' Public')

df['Workclass'] = df['Workclass'].replace([' Self-emp-not-inc',' Self-emp-inc'], ' Self-emp')

#df.shape

df['Workclass'].unique()


```

```python
df.isnull().sum()

df
```

```python

df1 = df.groupby(['Education','Attrtibutes'])


#total_count2.head()
summary_df1 = pd.DataFrame({"Education":df1['Education'].count(),"Salary":df1['Attrtibutes'].count()})
#summary_df1
del summary_df1['Education']

#summary_df1 = summary_df1.reset_index(level=[0,1])

#data = summary_df1.set_index(['Education', 'Attrtibutes']).Salary
#data

#data.unstack().plot(kind='bar', stacked=True)

summary_df1.unstack().plot(kind='bar', stacked=True)

#data.unstack().plot(kind='bar')

```

```python
df2 = df.groupby(['Marital_Status'])

hours_worked_per_group = df2['Hours per week'].sum()

total_hours = df['Hours per week'].sum()

#total_hours

hours_percentage_by_group = (hours_worked_per_group/total_hours)*100

hours_percentage_by_group

# Build Pie Chart

labels = ["Divorced", "Married", "Never-married", "Separated", "Widowed"]

sizes = [hours_percentage_by_group.loc[' Divorced'], hours_percentage_by_group.loc[' Married'], 
         hours_percentage_by_group.loc[' Never-married'], hours_percentage_by_group.loc[' Separated'], hours_percentage_by_group.loc[' Widowed']]

colors = ["lightcoral", "lightskyblue", "gold", "yellowgreen", "lightslategray"]

explode = (0, 0.1, 0, 0 , 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=280)
```
